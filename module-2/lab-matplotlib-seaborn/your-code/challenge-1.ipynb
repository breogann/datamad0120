#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# TIPO 1: crear un subplot por cada gráfico indicando su índice

plt.subplot(1,2,1)
plt.plot(x, y, color='red', ls="--")
plt.subplots_adjust(wspace = 0.2, right = 2)
plt.subplot(1,2,2)
plt.plot(x, z, ls='-', lw=5);
```

```python
# TIPO 2: crear un único subplot y configurar cada gráfico indicando antes la posición (mejor)

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y,lw=10)
ax[1].plot(x,z,color='g',ls='--')
fig.subplots_adjust(wspace = 0.2, right = 2)
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# TAMAÑO

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,5))

ax[0].plot(x,y,lw=10)
ax[1].plot(x,z,color='g',ls='--');
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# NOMBRE DE LOS EJES

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11,5))

ax[0].plot(x,y,lw=10)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].plot(x,z,color='g',ls='--')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y');
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# CAMBIAR LA ESCALA

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11,5))

ax[0].plot(x,x**2,x,np.exp(x),lw=10)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].plot(x,x**2,x, np.exp(x),color='g',ls='--')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
```
